```python
# silence warnings
import warnings
warnings.filterwarnings('ignore')
import yfinance as yf
import pyfolio as pf
%matplotlib inline
netflix = yf.Ticker('NFLX')
history= netflix.history('max')
history.index = history.index.tz_convert('UTC')
history.info()
```

```python
history.head()
```

```python
history.describe()
```

```python
# use pert_change() to calculate returns
returns = history.Close.pct_change()  
returns
```

```python
# create returns tear sheet 
pf.create_returns_tear_sheet(returns, live_start_date='2020-1-1')
```
